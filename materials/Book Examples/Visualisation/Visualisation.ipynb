```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
#import seaborn as sns
import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
actors = pd.read_csv('actors.csv')
actors.head()
```

```python
plt.scatter('Number of Movies', 'Total Gross', data=actors);
plt.xlabel('Number of Movies');
plt.ylabel('Total Gross');
```

```python
plt.scatter('Number of Movies', 'Average per Movie', data=actors)
plt.xlabel('Number of Movies')
plt.ylabel('Average per Movie')
```

```python
movie_by_year = pd.read_csv('movies_by_year.csv')
movie_by_year.head()
```

```python
plt.figure(figsize=[8, 4])
plt.plot('Year', 'Number of Movies', data=movie_by_year);
plt.xlabel('Year');
plt.ylabel('Number of Movies');
```

```python
century_21 = movie_by_year[movie_by_year['Year'] >= 2000]
century_21.head()
```

```python
plt.plot('Year', 'Number of Movies', data = century_21);
plt.xlabel('Year');
plt.ylabel('Number of Movies');
plt.xlim(2000, 2016);
```

```python
plt.plot('Year', 'Total Gross', data = century_21);
plt.xlabel('Year');
plt.ylabel('Total Gross');
plt.xlim(2000, 2016);
```
